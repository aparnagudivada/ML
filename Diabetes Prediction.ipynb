```python
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt
```

```python
dataset=load_diabetes()
```

```python
dataset
```

```python
dataset['target']
```

```python
dataset['feature_names']
```

```python
import pandas as pd
df=pd.DataFrame(dataset.data,columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
```

```python
df.head()
```

```python
x=df
y=dataset['target']
```

```python
## train test split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
## correlation
x_train.corr()
```

```python
import seaborn as sns
plt.figure(figsize=(15,10))
sns.heatmap(x_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor=DecisionTreeRegressor()
regressor.fit(x_train,y_train)
```

# Hyperparameter Tuning

```python
param={
    'criterion':['squared_error','friedman_mse','absolute_error'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['auto','sqrt','log2']
    
}
param
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
regressor=DecisionTreeRegressor()
```

```python
grid=GridSearchCV(regressor,param_grid=param,cv=5,scoring='neg_mean_squared_error')
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred1=grid.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print(r2_score(y_test,y_pred1))
```

```python
print(mean_absolute_error(y_test,y_pred1))
```

```python
print(mean_squared_error(y_test,y_pred1))
```

```python
selected= DecisionTreeRegressor(criterion='friedman_mse',max_depth=4,max_features='auto',splitter='random')
```

```python
selected.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(selected,filled=True)
```
